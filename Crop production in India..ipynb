```python
rice=data[[ 'rice area (1000 ha)', 'rice production (1000 tons)',
       'rice yield (kg per ha)','year']]
rice
rice.head()
```

```python
rice_year=rice.groupby('year')
rice_year.max()
rice_year.sum()
```

```python
#  year of max yeild of rice and production and sum  
rice_max=rice_year.sum().sort_values(by=['rice yield (kg per ha)','rice production (1000 tons)'],ascending=False)
rice_max
```

```python
rice_max_plot=rice_max[['rice production (1000 tons)','rice yield (kg per ha)']]
rice_max_plot
```

```python
plt.style.use('ggplot')
```

```python
# rice production over the years 
plt.plot(rice_max_plot['rice yield (kg per ha)'],marker='o',label='rice production')
plt.ylabel('rice yield')
plt.xlabel('years')
plt.title('rice production over the years')
plt.show
plt.legend()
```

```python
# Create a figure and a set of subplots
fig, ax = plt.subplots()

#plot rice production and rice yield
ax.plot(rice_max['rice production (1000 tons)'],label='Rice production(Kilotons)',marker='o',color='black',linestyle='')
ax.plot(rice_max['rice yield (kg per ha)'],label='Rice yield',marker='D',color='green',linestyle='')

# Set the title and labels
ax.set_title('Rice Yield and Production')
ax.set_xlabel('Years')
ax.set_ylabel('Value')

ax.legend()
```

```python
wheat=data[['wheat area (1000 ha)','wheat production (1000 tons)', 'wheat yield (kg per ha)','year']]
wheat.head(10)
```

```python
wheat.groupby('year').describe()
```

```python
wheat_max=wheat.groupby('year').max().sort_values(by='wheat production (1000 tons)',ascending=False)
wheat_max
```

```python
wheat_sum=wheat.groupby('year').sum().sort_values(by='wheat production (1000 tons)',ascending=False)
wheat_sum
```

```python
fig=plt.figure(figsize=(16,9),facecolor='cyan')

rows=1
columns=3

chart1=fig.add_subplot(rows,columns,1)
chart2=fig.add_subplot(rows,columns,2)
chart3=fig.add_subplot(rows,columns,3)

labels=['wheat area','wheat production','wheat yield']

chart1.plot(wheat_sum['wheat area (1000 ha)'],color='orange',label='wheat area',marker='*',linestyle='--',markersize=20)
chart2.plot(wheat_sum['wheat production (1000 tons)'],color='red',label='wheat production',marker='o',linestyle='-',markersize=15)
chart3.plot(wheat_sum['wheat yield (kg per ha)'],color='green',label='wheat yield',marker='D',linestyle='--',markersize=10)

plt.legend(labels)
```
